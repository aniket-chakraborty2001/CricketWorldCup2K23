```python
# Importing Required packages
import numpy as np
import pandas as pd
import folium
```

```python
# Creating the Indian Map Using Folium and Latitude and Longitude
india_map = folium.Map(location = [20.5937,78.9629], zoom_start = 4)
```

```python
# The locations of the Stadiums
places = [
    {'name': 'Kolkata', 'location': [22.5646995,88.3424417]},
    {'name': 'Ahmedabad', 'location': [23.0917717,72.5973345]},
    {'name': 'Hyderabad', 'location': [17.406503677368164,78.55042266845703]},
    {'name': 'Dharamsala', 'location' : [32.1975918,76.3250337]},
    {'name': 'Delhi', 'location': [28.6375309,77.2429345]},
    {'name': 'Chennai', 'location': [13.0627521,80.2784304]},
    {'name': 'Lucknow', 'location': [26.811283,81.0167051]},
    {'name': 'Pune', 'location': [18.6732563,73.7062647]},
    {'name': 'Bangaluru', 'location': [12.9788127,77.5995775]},
    {'name': 'Mumbai', 'location': [18.9385497,72.8257408]}
]

# Adding Markers and Place Names in Indian Map
for place in places:
  folium.Marker(
      location = place['location'],
      popup = folium.Popup(place['name'],parse_html = True),
      icon = folium.Icon(color = 'red')
  ).add_to(india_map)

# Getting the Complete Indian Map With Places
india_map
```
