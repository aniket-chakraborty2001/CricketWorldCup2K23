It is the next step of the project that deals with effective data visualization of different aspects to get meaningful insights and hidden patterns from the given data set.
```python
# Reading the 'WorldCup2K23_OriginalData.csv' file in a variable data
data = pd.read_csv('/content/drive/MyDrive/Project_Data_Sets/WorldCup2K23data_OriginalData.csv')
data.head()
```

```python
# Changing the index of the data set
# Now, the data is indexed by PlayerName
data_indexed = data.set_index('PlayerName')
data_indexed.head()
```

```python
# Droping some column to reduce the data set
# Drop the columns TeamName, Continent and Type
drop_col = ['TeamName','Continent','Type']
data_precise = data_indexed.drop(columns = drop_col, axis = 1)
data_precise.head()
```

```python
# Treating Missing Values
# Replace Missing Values with 0
data_precise.fillna(0, inplace = True)
data_precise.head(15)
```

```python
# Question-1: Visualize Runs Scored by Virat Kohli in 11 matches

# Getting the data for Virat Kohli
runsmatch = list(map(str,range(1,12)))
data_vk = data_precise.loc['Virat Kohli', runsmatch]

# Plotting graph for Virat Kohli
plt.figure(figsize = (6,4))
data_vk.plot(kind = 'line', color = 'red',marker = 's')
plt.xlabel('Number of Matches', fontsize = 14)
plt.ylabel('Runs Scored in Mathches', fontsize = 14)
plt.title('Runs Scored Trend of Virat Kohli In 2k23 WorldCup', fontsize = 14)
plt.grid(True)
plt.show()
```

```python
# Treating missing values in original data set
data.fillna(0, inplace = True)
data.head()
```

```python
# Question-14: % Contribution of Continets in Total Runs Scored
continentwise_run = data.groupby('Continent')['TotalRuns'].sum()
print(continentwise_run)
labels = continentwise_run.index
sizes = continentwise_run.values

# Plotting Pie Chart to represent % Contribution of Runs Scored
plt.pie(sizes,
        labels = labels,
        autopct = '%1.2f%%',
        colors = ['red','green','blue','yellow'],
        startangle = 90)
plt.title('% Contribution of Continents In Total Runs Scored')
plt.show()
```

```python
# Question-15: % Contribution of Nations in Total Runs Scored
nationwise_run = data.groupby('TeamName')['TotalRuns'].sum()
print(nationwise_run)
labels = nationwise_run.index
sizes = nationwise_run.values

# Plotting Pie Chart to represent % Contribution of Runs Scored
plt.pie(sizes,
        labels = labels,
        autopct = '%1.2f%%',
        colors = ['red','green','blue','yellow','orange','pink','grey',
                  'khaki','cyan','crimson'],
        startangle = 90)
plt.title('% Contribution of Continents In Total Runs Scored')
plt.show()
```

```python
# Question-16: % Contribution of Continets in Total Wickets Taken
continentwise_wicket = data.groupby('Continent')['WicketsTaken'].sum()
print(continentwise_wicket)
labels = continentwise_wicket.index
sizes = continentwise_wicket.values

# Plotting Pie Chart to represent % Contribution of Runs Scored
plt.pie(sizes,
        labels = labels,
        autopct = '%1.2f%%',
        colors = ['khaki','crimson','grey','cyan'],
        startangle = 90)
plt.title('% Contribution of Continents In Total Wickets Taken')
plt.show()
```
