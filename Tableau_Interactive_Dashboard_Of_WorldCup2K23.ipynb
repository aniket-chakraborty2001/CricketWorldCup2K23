## Tabluau Public:
Tabluau public is a bussiness intelligence (BI) tool used to create interactive visualizations. More popularly it is used to create a dashboard. A dashboard is a page/web page, where anyone can show the visualizations they make in the project. It is very helpful to understand the insights hidden in data. It is very efficient and time saving technique to create interactive visualizations.
